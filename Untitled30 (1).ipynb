Imports and Setup

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
```

 Data Acquisition


```python
tickers = ['AAPL', 'GOOGL', 'NVDA', 'MSFT', 'AMZN', 'JPM', 'MS', 'META', 'V', 'CRM']
data = yf.download(tickers, start="2015-01-01", end="2025-01-01", auto_adjust=False)
adj_close = data['Adj Close']

```

```python



adj_close = adj_close.interpolate(method='time').fillna(method='ffill', limit=5).dropna()

print("Data cleaned. Remaining NaNs:", adj_close.isnull().sum().sum())
```

Portfolio Construction


```python
import numpy as np

weights = np.array([0.1] * len(tickers))
daily_returns = adj_close.pct_change().dropna()
portfolio_returns = (daily_returns * weights).sum(axis=1)

```

```python

cumulative_return = (1 + portfolio_returns).cumprod() - 1
print("ðŸ“ˆ Cumulative Return (final value):")
print(cumulative_return.tail(1), "\n")

annualized_return = portfolio_returns.mean() * 252
print("ðŸ“ˆ Annualized Return:")
print(f"{annualized_return:.4f}\n")

annualized_volatility = portfolio_returns.std() * np.sqrt(252)
print("ðŸ“‰ Annualized Volatility:")
print(f"{annualized_volatility:.4f}\n")

sharpe_ratio = (annualized_return - 0.05) / annualized_volatility
print("ðŸ“Š Sharpe Ratio:")
print(f"{sharpe_ratio:.4f}")
```

```python

cumulative = (1 + portfolio_returns).cumprod()
rolling_max = cumulative.cummax()
drawdown = (cumulative - rolling_max) / rolling_max

max_drawdown = drawdown.min()
drawdown_duration = (drawdown != 0).astype(int).groupby((drawdown == 0).astype(int).cumsum()).sum().max()

print("ðŸ“‰ Max Drawdown:")
print(f"{max_drawdown:.4%}")
print("ðŸ“‰ Drawdown Duration (days):")
print(drawdown_duration)
```

```python

correlation_matrix = daily_returns.corr()
print("Correlation Matrix:")

print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cumulative.plot(title="Portfolio Cumulative Return", figsize=(10, 5))
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()

daily_returns.rolling(60).std().plot(figsize=(12, 6), title='Rolling Volatility (60-day)')
plt.show()

drawdown.plot(title="Portfolio Drawdown", figsize=(10, 5))
plt.show()
```

```python
from scipy.optimize import minimize


def neg_sharpe(weights, mean_returns, cov_matrix, risk_free_rate=0.05):
    port_return = np.dot(weights, mean_returns) * 252
    port_vol = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
    return -(port_return - risk_free_rate) / port_vol

mean_returns = daily_returns.mean()
cov_matrix = daily_returns.cov()
bounds = [(0, 1)] * len(tickers)
constraints = {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}

opt_result = minimize(neg_sharpe, weights, args=(mean_returns, cov_matrix), method='SLSQP',
                      bounds=bounds, constraints=constraints)

optimal_weights = opt_result.x

print(" Optimized Weights:")
for t, w in zip(tickers, optimal_weights):
    print(f"{t}: {w:.4f}")

opt_port_returns = (daily_returns * optimal_weights).sum(axis=1)
opt_annualized_return = opt_port_returns.mean() * 252
opt_annualized_volatility = opt_port_returns.std() * np.sqrt(252)
opt_sharpe = (opt_annualized_return - 0.05) / opt_annualized_volatility

print("\n Optimized Portfolio Performance:")
print(f"Annualized Return: {opt_annualized_return:.4f}")
print(f"Annualized Volatility: {opt_annualized_volatility:.4f}")
print(f"Sharpe Ratio: {opt_sharpe:.4f}")
```

```python
opt_portfolio_returns = (daily_returns * optimal_weights).sum(axis=1)

opt_cumulative = (1 + opt_portfolio_returns).cumprod()
opt_annualized_return = opt_portfolio_returns.mean() * 252
opt_annualized_volatility = opt_portfolio_returns.std() * np.sqrt(252)
opt_sharpe = (opt_annualized_return - risk_free_rate) / opt_annualized_volatility

print("Original Sharpe:", sharpe_ratio)
print("Optimized Sharpe:", opt_sharpe)

plt.figure(figsize=(10, 5))
cumulative.plot(label='Original Portfolio')
opt_cumulative.plot(label='Optimized Portfolio')
plt.legend()
plt.title("Original vs Optimized Portfolio Performance")
plt.grid(True)
plt.show()
```
